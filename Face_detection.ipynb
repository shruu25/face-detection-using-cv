FACE DETECTION of IMAGES

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
%matplotlib inline
```

LOADING IMAGES

```python
img_1 = cv2.imread('/baby_1.jpg')
img_1_grey = cv2.cvtColor(img_1,cv2.COLOR_BGR2GRAY)
plt.imshow(img_1_grey,cmap='gray')
```

```python
def convertToRGB(img):
  return cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

LOADING HAAR CASCADE CLASSIFIER

```python
haar_cascade_face = cv2.CascadeClassifier('/haarcascade_frontalface_alt2.xml')
```

```python
faces_rects = haar_cascade_face.detectMultiScale(img_1_grey, scaleFactor = 1.2, minNeighbors = 5);
print('Faces found: ', len(faces_rects))
```

```python
for (x,y,w,h) in faces_rects:
     cv2.rectangle(img_1, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
plt.imshow(convertToRGB(img_1))
```

```python
def detect_faces(cascade,img_1, scaleFactor = 1.1):
    # create a copy of the image to prevent any changes to the original one.
    image_copy = img_1.copy()
    
    #convert the test image to gray scale as opencv face detector expects gray images
    gray_image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)
    
    # Applying the haar classifier to detect faces
    faces_rect = cascade.detectMultiScale(gray_image, scaleFactor=scaleFactor, minNeighbors=5)
    
    for (x, y, w, h) in faces_rect:
        cv2.rectangle(image_copy, (x, y), (x+w, y+h), (0, 255, 0), 15)
        
    return image_copy
```

second image for face detection

```python
img_2 = cv2.imread('/babyface.jpg')

#call the function to detect faces
faces = detect_faces(haar_cascade_face, img_2)

#convert to RGB and display image
plt.imshow(convertToRGB(faces))
```

for group pictures face detection

```python
img_3 = cv2.imread('/school.jpg')

#call the function to detect faces
faces = detect_faces(haar_cascade_face, img_3)

#convert to RGB and display image
plt.imshow(convertToRGB(faces))
```
